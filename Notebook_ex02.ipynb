```python
import pandas as pd
df = pd.DataFrame({
    'name':['christopher','marion','maria','mia','clement','randy','remi'],
    'age':[70,30,22,19,45,33,20],
    'num_children':[4,2,1,0,3,1,0],
    'num_pets':[5,1,0,2,2,2,3]
})
df.plot(x="age", y="num_children", kind="scatter", title="Age vs Children", xlabel="Age", ylabel="Children")
```
