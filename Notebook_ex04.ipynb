```python
import matplotlib.pyplot as plt

left=[5,7,11,13,17]
right=[0.1,0.2,0.4,0.8,-1.6]
x=[0,1,2,3,4]

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x,left,"k-",label="Left")  # black
ax2.plot(x,right,"r-",label="Right") # red
ax1.set_ylabel("Left values"); ax2.set_ylabel("Right values")
plt.title("Two Axes Plot")
plt.show()
```
