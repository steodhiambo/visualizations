```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[1,4,2,5,7,1,6,3]

plt.plot(x,y,"r-.",linewidth=3)             # red dashdot line
plt.scatter(x,y,c="b",s=12**2)             # blue circles size 12
plt.title("Matplotlib Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(1,8); plt.ylim(1,8)
plt.show()
```
